This script plots the decay of average displacements measured in the field with fault-perpendicular distance for the Landers, Hector Mine, El Mayor Cucapah, and Ridgecrest earthquakes. The decays are fit with an ensemble sampler for MCMC (see separate scripts) and the posterior distributions for each parameter shown in the last cell of this notebook.

### Load data and plot average displacement decay

```python
import numpy as np
import matplotlib.pyplot as plt
# load Landers data
Landers_decay = np.loadtxt('Landers_mean_displacement_decay.txt', delimiter=',')
Landers_displacement = Landers_decay[:,1]
Landers_fit = np.loadtxt('Landers_MCMCfits_displacement.txt', delimiter=' ')
Landers_spread = np.loadtxt('Landers_spread_displacement.txt', delimiter=' ')
xLanders = Landers_fit[:,0]
bestfit_Landers = Landers_fit[:,1]
med_Landers = Landers_spread[:,0]
spread_Landers = Landers_spread[:,1]
best_fit_parameters_Landers = np.loadtxt('Landers_best_fit_parameters_displacement.txt', delimiter=' ')
Landers_g = best_fit_parameters_Landers[1] # n
Landers_d = best_fit_parameters_Landers[0] # beta


# load EMC data
EMC_decay = np.loadtxt('EMC_mean_displacement_decay.txt', delimiter=',')
EMC_displacement = EMC_decay[:,1]
EMC_fit = np.loadtxt('EMC_MCMCfits_displacement.txt', delimiter=' ')
EMC_spread = np.loadtxt('EMC_spread_displacement.txt', delimiter=' ')
xEMC = EMC_fit[:,0]
bestfit_EMC = EMC_fit[:,1]
med_EMC = EMC_spread[:,0]
spread_EMC = EMC_spread[:,1]
best_fit_parameters_EMC = np.loadtxt('EMC_best_fit_parameters_displacement.txt', delimiter=' ')
EMC_g = best_fit_parameters_EMC[1]
EMC_d = best_fit_parameters_EMC[0]


# load Hector Mine data
HectorMine_decay = np.loadtxt('HectorMine_mean_displacement_decay.txt', delimiter=',')
HectorMine_displacement = HectorMine_decay[:,1]
HectorMine_fit = np.loadtxt('HectorMine_MCMCfits_displacement.txt', delimiter=' ')
HectorMine_spread = np.loadtxt('HectorMine_spread_displacement.txt', delimiter=' ')
xHectorMine = HectorMine_fit[:,0]
bestfit_HectorMine = HectorMine_fit[:,1]
med_HectorMine = HectorMine_spread[:,0]
spread_HectorMine = HectorMine_spread[:,1]
best_fit_parameters_HectorMine = np.loadtxt('HectorMine_best_fit_parameters_displacement.txt', delimiter=' ')
HectorMine_g = best_fit_parameters_HectorMine[1]
HectorMine_d = best_fit_parameters_HectorMine[0]


# load Ridgecrest1 data
Ridgecrest1_decay = np.loadtxt('Ridgecrest1_mean_displacement_decay.txt', delimiter=',')
Ridgecrest1_displacement = Ridgecrest1_decay[:,1]
Ridgecrest1_fit = np.loadtxt('Ridgecrest1_MCMCfits_displacement.txt', delimiter=' ')
Ridgecrest1_spread = np.loadtxt('Ridgecrest1_spread_displacement.txt', delimiter=' ')
xRidgecrest1 = Ridgecrest1_fit[:,0]
bestfit_Ridgecrest1 = Ridgecrest1_fit[:,1]
med_Ridgecrest1 = Ridgecrest1_spread[:,0]
spread_Ridgecrest1 = Ridgecrest1_spread[:,1]
best_fit_parameters_Ridgecrest1 = np.loadtxt('Ridgecrest1_best_fit_parameters_displacement.txt', delimiter=' ')
Ridgecrest1_g = best_fit_parameters_Ridgecrest1[1]
Ridgecrest1_d = best_fit_parameters_Ridgecrest1[0]


# load Ridgecrest2 data 
Ridgecrest2_decay = np.loadtxt('Ridgecrest2_mean_displacement_decay.txt', delimiter=',')
Ridgecrest2_displacement = Ridgecrest2_decay[:,1]
Ridgecrest2_fit = np.loadtxt('Ridgecrest2_MCMCfits_displacement.txt', delimiter=' ')
Ridgecrest2_spread = np.loadtxt('Ridgecrest2_spread_displacement.txt', delimiter=' ')
xRidgecrest2= Ridgecrest2_fit[:,0]
bestfit_Ridgecrest2= Ridgecrest2_fit[:,1]
med_Ridgecrest2= Ridgecrest2_spread[:,0]
spread_Ridgecrest2= Ridgecrest2_spread[:,1]
best_fit_parameters_Ridgecrest2= np.loadtxt('Ridgecrest2_best_fit_parameters_displacement.txt', delimiter=' ')
Ridgecrest2_g = best_fit_parameters_Ridgecrest2[1]
Ridgecrest2_d = best_fit_parameters_Ridgecrest2[0]
```

```python
fig, axs = plt.subplots(1,5, dpi=300, figsize = (12,3), tight_layout = True, sharex=True)

axs[0].scatter(xLanders,Landers_displacement,color='firebrick')
axs[0].plot(xLanders,bestfit_Landers,label='Landers',color='firebrick')
axs[0].fill_between(xLanders,med_Landers-spread_Landers,med_Landers+spread_Landers,color='firebrick',alpha=0.2)
axs[0].set_yscale('log')
axs[0].set_xscale('log')
axs[0].set_ylabel('Mean displacement (m)')
axs[0].set_xlabel('Fault-perpendicular distance (m)')

axs[1].scatter(xHectorMine,HectorMine_displacement,color='darkseagreen')
axs[1].plot(xHectorMine,bestfit_HectorMine,label=' Hector Mine',color='darkseagreen')
axs[1].fill_between(xHectorMine,med_HectorMine-spread_HectorMine,med_HectorMine+spread_HectorMine,color='darkseagreen',alpha=0.2)
axs[1].set_yscale('log')
axs[1].set_xscale('log')
axs[1].set_ylabel('Mean displacement (m)')
axs[1].set_xlabel('Fault-perpendicular distance (m)')

axs[2].scatter(xEMC,EMC_displacement,color='darkcyan')
axs[2].plot(xEMC,bestfit_EMC,label='EMC',color='darkcyan')
axs[2].fill_between(xEMC,med_EMC-spread_EMC,med_EMC+spread_EMC,color='darkcyan',alpha=0.2)
axs[2].set_yscale('log')
axs[2].set_xscale('log')
axs[2].set_ylabel('Mean displacement (m)')
axs[2].set_xlabel('Fault-perpendicular distance (m)')

axs[3].scatter(xRidgecrest2,Ridgecrest2_displacement,color='purple')
axs[3].plot(xRidgecrest2,bestfit_Ridgecrest2,label='Ridgecrest mainshock',color='purple')
axs[3].fill_between(xRidgecrest2,med_Ridgecrest2-spread_Ridgecrest2,med_Ridgecrest2+spread_Ridgecrest2,color='purple',alpha=0.2)
axs[3].set_yscale('log')
axs[3].set_xscale('log')
axs[3].set_ylabel('Mean displacement (m)')
axs[3].set_xlabel('Fault-perpendicular distance (m)')

axs[4].scatter(xRidgecrest1,Ridgecrest1_displacement,color='orange')
axs[4].plot(xRidgecrest1,bestfit_Ridgecrest1,label='Ridgecrest foreshock',color='orange')
axs[4].fill_between(xRidgecrest1,med_Ridgecrest1-spread_Ridgecrest1,med_Ridgecrest1+spread_Ridgecrest1,color='orange',alpha=0.2)

axs[4].set_yscale('log')
axs[4].set_xscale('log')
axs[4].set_ylabel('Mean displacement (m)')
axs[4].set_xlabel('Fault-perpendicular distance (m)')

plt.tight_layout()
plt.savefig('all_event_displacement_decays.pdf')
plt.show()
```

### load posterior distributions for $x_S$ and $\textit{n}$ and plot them

```python
# distribution of parameters
parameters_Ridgecrest1 = np.loadtxt('Ridgecrest1_parameters_displacement.txt', delimiter=' ')
parameters_Ridgecrest2 = np.loadtxt('Ridgecrest2_parameters_displacement.txt', delimiter=' ')
parameters_EMC = np.loadtxt('EMC_parameters_displacement.txt', delimiter =' ')
parameters_HectorMine = np.loadtxt('HectorMine_parameters_displacement.txt', delimiter=' ')
parameters_Landers = np.loadtxt('Landers_parameters_displacement.txt', delimiter=' ')
```

```python
# histogram with parameter distributions 
# gamma
plt.figure()
fig, axs = plt.subplots(2,5, dpi=300, tight_layout = True)
n_bins = 100
axs[0,0].hist(parameters_Ridgecrest1[:,0], bins=n_bins,orientation='horizontal',color='orange',alpha=0.5)
axs[1,0].hist(parameters_Ridgecrest1[:,1], bins=n_bins,orientation='horizontal',color='orange',alpha=0.5)
axs[0,0].axhline(y=Ridgecrest1_d,color='orange',linewidth=0.5)
axs[1,0].axhline(y=Ridgecrest1_g,color='orange',linewidth=0.5)
axs[1,0].set_ylabel('$n$')
axs[0,0].set_ylabel('')
axs[0,1].hist(parameters_Ridgecrest2[:,0], bins=n_bins,orientation='horizontal',color='purple',alpha=0.5)
axs[0,1].axhline(y=Ridgecrest2_d,color='purple',linewidth=0.5)
axs[1,1].axhline(y=Ridgecrest2_g,color='purple',linewidth=0.5)
axs[1,1].hist(parameters_Ridgecrest2[:,1], bins=n_bins,orientation='horizontal',color='purple',alpha=0.5)
axs[0,2].hist(parameters_EMC[:,0], bins=n_bins,orientation='horizontal',color='darkcyan',alpha=0.5)
axs[0,2].axhline(y=EMC_d,color='darkcyan',linewidth=0.5)
axs[1,2].hist(parameters_EMC[:,1], bins=n_bins,orientation='horizontal',color='darkcyan',alpha=0.5)
axs[1,2].axhline(y=EMC_g,color='darkcyan',linewidth=0.5)
axs[0,4].hist(parameters_HectorMine[:,0], bins=n_bins,orientation='horizontal',color='darkseagreen',alpha=0.5)
axs[0,4].axhline(y=HectorMine_d,color='darkseagreen',linewidth=0.5)
axs[1,4].hist(parameters_HectorMine[:,1], bins=n_bins,orientation='horizontal',color='darkseagreen',alpha=0.5)
axs[1,4].axhline(y=HectorMine_g,color='darkseagreen',linewidth=0.5)
axs[0,3].hist(parameters_Landers[:,0], bins=n_bins,orientation='horizontal',color='firebrick',alpha=0.5)
axs[0,3].axhline(y=Landers_d,color='firebrick',linewidth=0.5)
axs[1,3].hist(parameters_Landers[:,1], bins=n_bins,orientation='horizontal',color='firebrick',alpha=0.5)
axs[1,3].axhline(y=Landers_g,color='firebrick',linewidth=0.5)


axs[0,0].axis(ymin=0,ymax= 15)
axs[0,1].axis(ymin=0,ymax= 15)
axs[0,2].axis(ymin=0,ymax= 15)
axs[0,3].axis(ymin=0,ymax= 15)
axs[0,4].axis(ymin=0,ymax= 15)
axs[1,0].axis(ymin=0,ymax=1)
axs[1,1].axis(ymin=0,ymax=1)
axs[1,2].axis(ymin=0,ymax=1)
axs[1,3].axis(ymin=0,ymax=1)
axs[1,4].axis(ymin=0,ymax=1)


for x in range(0,5):
    axs[0,x].set_xticklabels([])
    axs[1,x].set_xticklabels([])

for x in range(1,5):
    axs[0,x].set_yticklabels([])
    axs[1,x].set_yticklabels([])
    axs[0,x].axis('off')
    axs[1,x].axis('off')
plt.savefig('posterior_distribution_PFDHA_allevents_displacement.pdf')
plt.show()
```
