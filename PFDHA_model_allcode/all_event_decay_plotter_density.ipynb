This script plots the decay of fracture density with fault-perpendicular distance for the Landers, Hector Mine, El Mayor Cucapah, and Ridgecrest earthquakes. The decays are fit with an ensemble sampler for MCMC (see separate scripts) and the posterior distributions for each parameter shown in the last cell of this notebook.

### Load data and plot fracture density decay

```python
import numpy as np
import matplotlib.pyplot as plt
# load Landers data
Landers_decay = np.loadtxt('Landers_density_decay.txt', delimiter=',')
Landers_density = Landers_decay[:,1]
Landers_fit = np.loadtxt('Landers_MCMCfits.txt', delimiter=' ')
Landers_spread = np.loadtxt('Landers_spread.txt', delimiter=' ')
xLanders = Landers_fit[:,0]
bestfit_Landers = Landers_fit[:,1]
med_Landers = Landers_spread[:,0]
spread_Landers = Landers_spread[:,1]
best_fit_parameters_Landers = np.loadtxt('Landers_best_fit_parameters.txt', delimiter=' ')
Landers_g = best_fit_parameters_Landers[2]
Landers_d = best_fit_parameters_Landers[1]

# load EMC data
EMC_decay = np.loadtxt('EMC_density_decay.txt', delimiter=',')
EMC_density = EMC_decay[:,1]
EMC_fit = np.loadtxt('EMC_MCMCfits.txt', delimiter=' ')
EMC_spread = np.loadtxt('EMC_spread.txt', delimiter=' ')
xEMC = EMC_fit[:,0]
bestfit_EMC = EMC_fit[:,1]
med_EMC = EMC_spread[:,0]
spread_EMC = EMC_spread[:,1]
best_fit_parameters_EMC = np.loadtxt('EMC_best_fit_parameters.txt', delimiter=' ')
EMC_g = best_fit_parameters_EMC[2]
EMC_d = best_fit_parameters_EMC[1]

# load Hector Mine data
HectorMine_decay = np.loadtxt('HectorMine_density_decay.txt', delimiter=',')
HectorMine_density = HectorMine_decay[:,1]
HectorMine_fit = np.loadtxt('HectorMine_MCMCfits.txt', delimiter=' ')
HectorMine_spread = np.loadtxt('HectorMine_spread.txt', delimiter=' ')
xHectorMine = HectorMine_fit[:,0]
bestfit_HectorMine = HectorMine_fit[:,1]
med_HectorMine = HectorMine_spread[:,0]
spread_HectorMine = HectorMine_spread[:,1]
best_fit_parameters_HectorMine = np.loadtxt('HectorMine_best_fit_parameters.txt', delimiter=' ')
HectorMine_g = best_fit_parameters_HectorMine[2]
HectorMine_d = best_fit_parameters_HectorMine[1]

# load Ridgecrest1 data
Ridgecrest1_decay = np.loadtxt('Ridgecrest1_density_decay.txt', delimiter=',')
Ridgecrest1_density = Ridgecrest1_decay[:,1]
Ridgecrest1_fit = np.loadtxt('Ridgecrest1_MCMCfits.txt', delimiter=' ')
Ridgecrest1_spread = np.loadtxt('Ridgecrest1_spread.txt', delimiter=' ')
xRidgecrest1 = Ridgecrest1_fit[:,0]
bestfit_Ridgecrest1 = Ridgecrest1_fit[:,1]
med_Ridgecrest1 = Ridgecrest1_spread[:,0]
spread_Ridgecrest1 = Ridgecrest1_spread[:,1]
best_fit_parameters_Ridgecrest1 = np.loadtxt('Ridgecrest1_best_fit_parameters.txt', delimiter=' ')
Ridgecrest1_g = best_fit_parameters_Ridgecrest1[2]
Ridgecrest1_d = best_fit_parameters_Ridgecrest1[1]

# load Ridgecrest2 data 
Ridgecrest2_decay = np.loadtxt('Ridgecrest2_density_decay.txt', delimiter=',')
Ridgecrest2_density = Ridgecrest2_decay[:,1]
Ridgecrest2_fit = np.loadtxt('Ridgecrest2_MCMCfits.txt', delimiter=' ')
Ridgecrest2_spread = np.loadtxt('Ridgecrest2_spread.txt', delimiter=' ')
xRidgecrest2= Ridgecrest2_fit[:,0]
bestfit_Ridgecrest2= Ridgecrest2_fit[:,1]
med_Ridgecrest2= Ridgecrest2_spread[:,0]
spread_Ridgecrest2= Ridgecrest2_spread[:,1]
best_fit_parameters_Ridgecrest2= np.loadtxt('Ridgecrest2_best_fit_parameters.txt', delimiter=' ')
Ridgecrest2_g = best_fit_parameters_Ridgecrest2[2]
Ridgecrest2_d = best_fit_parameters_Ridgecrest1[1]
```

```python
best_fit_parameters_Ridgecrest2
```

```python
plt.figure()
plt.plot(xLanders,Landers_density,color='firebrick',alpha=0.5)
plt.plot(xLanders,bestfit_Landers,label='Landers',color='firebrick')
plt.fill_between(xLanders,med_Landers-spread_Landers,med_Landers+spread_Landers,color='firebrick',alpha=0.2)

plt.plot(xHectorMine,HectorMine_density,color='darkseagreen',alpha=0.5)
plt.plot(xHectorMine,bestfit_HectorMine,label=' Hector Mine',color='darkseagreen')
plt.fill_between(xHectorMine,med_HectorMine-spread_HectorMine,med_HectorMine+spread_HectorMine,color='darkseagreen',alpha=0.2)

plt.plot(xEMC,EMC_density,color='darkcyan',alpha=0.5)
plt.plot(xEMC,bestfit_EMC,label='EMC',color='darkcyan')
plt.fill_between(xEMC,med_EMC-spread_EMC,med_EMC+spread_EMC,color='darkcyan',alpha=0.2)

plt.plot(xRidgecrest2,Ridgecrest2_density,color='purple',alpha=0.5)
plt.plot(xRidgecrest2,bestfit_Ridgecrest2,label='Ridgecrest mainshock',color='purple')
plt.fill_between(xRidgecrest2,med_Ridgecrest2-spread_Ridgecrest2,med_Ridgecrest2+spread_Ridgecrest2,color='purple',alpha=0.2)

plt.plot(xRidgecrest1,Ridgecrest1_density,color='orange',alpha=0.5)
plt.plot(xRidgecrest1,bestfit_Ridgecrest1,label='Ridgecrest foreshock',color='orange')
plt.fill_between(xRidgecrest1,med_Ridgecrest1-spread_Ridgecrest1,med_Ridgecrest1+spread_Ridgecrest1,color='orange',alpha=0.2)

plt.yscale('log')
plt.xscale('log')
plt.ylabel('Fracture density ($n/m^{2})$')
plt.xlabel('Fault-perpendicular distance (m)')
plt.xlim(0,35000)
plt.legend()
plt.savefig('all_event_decays.pdf')
plt.show()
```

### load posterior distributions for $x_f$ and $\gamma$ and plot them

```python
# distribution of parameters
parameters_Ridgecrest1 = np.loadtxt('Ridgecrest1_parameters.txt', delimiter=' ')
parameters_Ridgecrest2 = np.loadtxt('Ridgecrest2_parameters.txt', delimiter=' ')
parameters_EMC = np.loadtxt('EMC_parameters.txt', delimiter =' ')
parameters_HectorMine = np.loadtxt('HectorMine_parameters.txt', delimiter=' ')
parameters_Landers = np.loadtxt('Landers_parameters.txt', delimiter=' ')
```

```python
# histogram with parameter distributions 
# gamma
plt.figure()
fig, axs = plt.subplots(2,5, dpi=300, tight_layout = True)
n_bins = 100
axs[0,0].hist(parameters_Ridgecrest1[:,1], bins=n_bins,orientation='horizontal',color='orange',alpha=0.5)
axs[1,0].hist(parameters_Ridgecrest1[:,2], bins=n_bins,orientation='horizontal',color='orange',alpha=0.5)
axs[0,0].axhline(y=Ridgecrest1_d,color='orange',linewidth=0.5)
axs[1,0].axhline(y=Ridgecrest1_g,color='orange',linewidth=0.5)
axs[1,0].set_ylabel('$\gamma$')
axs[0,0].set_ylabel('$x_{fr}$')
axs[0,1].hist(parameters_Ridgecrest2[:,1], bins=n_bins,orientation='horizontal',color='purple',alpha=0.5)
axs[0,1].axhline(y=Ridgecrest2_d,color='purple',linewidth=0.5)
axs[1,1].axhline(y=Ridgecrest2_g,color='purple',linewidth=0.5)
axs[1,1].hist(parameters_Ridgecrest2[:,2], bins=n_bins,orientation='horizontal',color='purple',alpha=0.5)
axs[0,2].hist(parameters_EMC[:,1], bins=n_bins,orientation='horizontal',color='darkcyan',alpha=0.5)
axs[0,2].axhline(y=EMC_d,color='darkcyan',linewidth=0.5)
axs[1,2].hist(parameters_EMC[:,2], bins=n_bins,orientation='horizontal',color='darkcyan',alpha=0.5)
axs[1,2].axhline(y=EMC_g,color='darkcyan',linewidth=0.5)
axs[0,4].hist(parameters_HectorMine[:,1], bins=n_bins,orientation='horizontal',color='darkseagreen',alpha=0.5)
axs[0,4].axhline(y=HectorMine_d,color='darkseagreen',linewidth=0.5)
axs[1,4].hist(parameters_HectorMine[:,2], bins=n_bins,orientation='horizontal',color='darkseagreen',alpha=0.5)
axs[1,4].axhline(y=HectorMine_g,color='darkseagreen',linewidth=0.5)
axs[0,3].hist(parameters_Landers[:,1], bins=n_bins,orientation='horizontal',color='firebrick',alpha=0.5)
axs[0,3].axhline(y=Landers_d,color='firebrick',linewidth=0.5)
axs[1,3].hist(parameters_Landers[:,2], bins=n_bins,orientation='horizontal',color='firebrick',alpha=0.5)
axs[1,3].axhline(y=Landers_g,color='firebrick',linewidth=0.5)


axs[0,0].axis(ymin=0,ymax=30)
axs[0,1].axis(ymin=0,ymax=30)
axs[0,2].axis(ymin=0,ymax=30)
axs[0,3].axis(ymin=0,ymax=30)
axs[0,4].axis(ymin=0,ymax=30)
axs[1,0].axis(ymin=0.5,ymax=2)
axs[1,1].axis(ymin=0.5,ymax=2)
axs[1,2].axis(ymin=0.5,ymax=2)
axs[1,3].axis(ymin=0.5,ymax=2)
axs[1,4].axis(ymin=0.5,ymax=2)


for x in range(0,5):
    axs[0,x].set_xticklabels([])
    axs[1,x].set_xticklabels([])

for x in range(1,5):
    axs[0,x].set_yticklabels([])
    axs[1,x].set_yticklabels([])
    axs[0,x].axis('off')
    axs[1,x].axis('off')
plt.savefig('gamma_distribution_PFDHA_allevents.pdf')
plt.show()
```
